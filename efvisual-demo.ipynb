# Particle moving with constant velocity

# Particle Moving in a circle, small $\beta$

```python
p = particle(1, ["cos(0.01*t)","sin(0.01*t)"])
rf = reference_frame(3, 20)
rf.anim_video(p, 800, "circle_0.01_beta")
```

# Demonstrating Bremsstrahlung radiation

Charge is decelerated by a tan-hyperbolic function.

```python
p = particle(1, ["1-tanh(0.9*t)","0"])
rf = reference_frame(3, 20)
rf.anim_video(p, 4, "bremsstrahlung_tanh_0.9_beta")
```
